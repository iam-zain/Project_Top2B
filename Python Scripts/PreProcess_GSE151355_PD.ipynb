```python
import csv
import pandas as pd

%cd Z:\Project_Top2B\April_2023\Files_Top2B\GSE151355

df2 = pd.read_csv('GSE151355_Age_PD.csv')
df2.head(2)
```

```python
# Removing P val columns
# renaming columns

df = pd.read_csv('GSE151355.csv')
#df = df[df.columns.drop(list(df.filter(regex='Pval')))]
#df.columns = df.columns.str.replace('_Beta', '')
print(df.shape)
df.head(2)
```

```python
# remove quotes from column names
df.columns = df.columns.str.replace(' ', '')

# remove quotes from cell values
df = df.applymap(lambda x: x.replace('"', '') if isinstance(x, str) else x)

# print updated DataFrame
print(df.shape)
df.head(2)
```

```python

# Extract the names which are control
names = df2['Sample_ID'].tolist()

# Subset df1 using the names extracted from df2
df1_subset = df[names]

# Adding ID_Ref column from df
df1_subset.insert(loc=0, column='ID_REF', value=df.iloc[:, 0])

df1_subset.to_csv('GSE151355_PD.csv', index = False)

print(df1_subset.shape)
df1_subset.head(2)
```

```python
%cd Z:\Project_Top2B
df1 = pd.read_csv('CpGLists_Top2B.csv')

%cd Z:\Project_Top2B\April_2023\Files_Top2B\GSE151355
df2 = pd.read_csv('GSE151355_PD.csv')

print(df1.head(2))
df2.head(2)
```
