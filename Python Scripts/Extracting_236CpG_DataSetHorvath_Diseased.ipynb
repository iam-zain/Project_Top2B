## Schizophrenia (GSE74193)

```python
import pandas as pd
%cd Z:\Project_Top2B\April_2023\diseasewise\schiz
    
dfh = pd.read_csv('236CpG_List.csv')
df1 = pd.read_csv('GSE74193_Top2b_Schiz_Trans.csv')
print(df1.shape)

df1m = pd.merge(dfh, df1, on = 'CpG', how = 'inner')
print(df1m.shape)
df1m.to_csv ('GSE74193_236CpG_Schiz.csv', index = False)
df1m.head(2)
```
