```python
import numpy as np
import pandas as pd

%cd Z:\Project_Top2B\April_2023\regionwise\cerebellum
data = pd.read_csv('MergedRO_Top2B_Cerebellum1.csv')
data = data.drop(['Sample_ID'], axis = 1)
data.head(1)
```

```python
df = pd.read_csv('List_ModelCpG_cb.csv')

# Get the unique values from the first column of df
columns_to_select = df.iloc[:, 0].unique()

# Select only the columns from data that are present in columns_to_select
selected_columns = data[columns_to_select]

# Add the column Age from data to the firstmost column of selected_columns
Age = 'Age'
selected_columns.insert(0, Age, data['Age'])

print(selected_columns.shape)
selected_columns.head(2)
```

```python
# Replace '' values with NaN
data = data.replace('', np.nan)

# Loop over each column and replace NaN values with column means
for col in data.columns:
    # Convert the column to numeric data type
    data[col] = pd.to_numeric(data[col])
    
    # Calculate the mean of the column
    col_mean = data[col].mean()
    
    # Replace NaN values with the column mean
    data[col].fillna(col_mean, inplace=True)

data.to_csv("MergedRO_Top2B_Mean_cb.csv", index = False)
data.head(2)
```
