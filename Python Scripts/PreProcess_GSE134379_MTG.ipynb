```python
import csv
import pandas as pd

%cd Z:\Project_Top2B\April_2023\GSE134379

# Define the input and output file paths
input_file = "GSE134379_processedSamples_mtgND.txt"
output_file = "GSE134379_mtg.csv"

# Read the text file
with open(input_file, 'r') as f:
    lines = f.readlines()

# Create a list of lists, where each sublist represents a row in the CSV file
data = [line.strip().split() for line in lines]

# Write the data to the CSV file
with open(output_file, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
# Removing P val columns
# renaming columns

df = pd.read_csv('GSE134379_mtg.csv')
df = df[df.columns.drop(list(df.filter(regex='Value')))]
df.columns = df.columns.str.replace('Beta_value_for_', '')
print(df.shape)
df.head(2)
```

```python
df2 = pd.read_csv('GSE134379_Age_MTG.csv')
df2.head(2)
```

```python
# Extract the names which are present in both df2 and df
names = list(set(df2['Sample']).intersection(set(df.columns)))

# Subset df using the names extracted from df2
df_subset = df[names]

# Adding CpG column from df
df_subset.insert(loc=0, column='CpG', value=df.iloc[:, 0])

df_subset.to_csv('GSE134379_mtg_noPval.csv', index=False)

print(df_subset.shape)
df_subset.head(2)
```
