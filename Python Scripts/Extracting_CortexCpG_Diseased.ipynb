## Schizophrenia (GSE74193)

```python
import pandas as pd
%cd Z:\Project_Top2B\April_2023\Individual_Files\GSE74193
    
dfh = pd.read_csv('CpG_347.csv')
df1 = pd.read_csv('GSE74193_Schiz.csv')
print(df1.shape)

df1m = pd.merge(dfh, df1, on = 'ID_REF', how = 'inner')
print(df1m.shape)
df1m.to_csv ('GSE74193_Schiz_Cortex.csv', index = False)
df1m.head(2)
```
