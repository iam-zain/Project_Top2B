```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%cd Z:\Project_Top2B\April_2023\Files_Top2B\CpG_inRow
```

```python
df = pd.read_csv('CpG_236BorutaPy_onAll.csv')
#df = df.drop(['PATNO', 'Patient_ID'], axis=1)
print(df.shape)
df.head(2)
```

```python
corrmat = df.corr()
top_corr_feat = corrmat.index
plt.figure(figsize=(40,30))
hm = sns.heatmap(df[top_corr_feat].corr().round(2), annot=True, cmap='RdYlGn')
hm.set_xticklabels(hm.get_xticklabels(), fontsize=16, fontweight='bold')
hm.set_yticklabels(hm.get_yticklabels(), fontsize=16, fontweight='bold')
plt.savefig("tempout_pat.png")
```
