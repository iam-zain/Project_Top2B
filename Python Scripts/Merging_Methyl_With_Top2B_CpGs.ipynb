```python
import pandas as pd
import numpy as np
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
from sklearn import preprocessing
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn import metrics
import seaborn as sns
```

```python
%cd Z:\PPMI_Data\Excels\Raw_And_More
df = pd.read_csv ("beta_post_Funnorm_PPMI_EPICn524final030618.csv")
print(df.shape)
df.head(2)
```

```python
df.index.name ='CpG'
df.reset_index(inplace=True)
df.head(2)
```

#Top2B CPGs

```python
%cd Z:\Project_Top2B\Clocks_Analysis\PD_PPMI
df_Top = pd.read_csv ("CpGLists_Top2B.csv")
print(df_Top.shape)
df_Top.head(2)
```

```python
df_TopM = pd.merge(df, df_Top, on='CpG', how='inner')
print(df_TopM.shape)
df_TopM.head(2)
```

```python
dft = df_TopM.transpose()
dft.head(2)
```

```python
dft.rename(columns=dft.iloc[0], inplace = True)
```

```python
dft.index.name ='CpG'
dft.reset_index(inplace=True)
dft.head(2)
```

```python
dft = dft.drop ([0], axis = 0)
dft.head(2)
```

```python
dft.rename (columns = {"CpG" : "Sentrix"}, inplace = True)
dft.head(2)
```

```python
df2 = pd.read_csv ("Sentrix_Age_Methyl.csv")
df2.head(2)
```

```python
df3 = pd.merge (df2, dft, on= "Sentrix", how= "inner")
print(df3.shape)
df3.head(2)
```

```python
df_data = df3.drop(['Sentrix'], axis=1)
df_data.head(2)
```

```python
df_data.to_csv('PPMI_Top2B.csv', index=False)
```
