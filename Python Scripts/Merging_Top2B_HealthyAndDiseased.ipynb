```python
import pandas as pd

%cd Z:\Project_Top2B\April_2023\all

df1 = pd.read_csv('Merged18_Top2B_Mean.csv')
df2 = pd.read_csv('Merged_Top2B_Diseased_Mean.csv')

df1 = df1.transpose()
df1.columns = df1.iloc[0]
df1 = df1.iloc [1:, :]

df2 = df2.transpose()
df2.columns = df2.iloc[0]
df2 = df2.iloc [1:, :]

df2.head(2)
```

```python
from functools import reduce

df_list = [df1, df2]

df_merged = reduce(lambda left, right: pd.merge(left, right, left_index=True, right_index=True, how='inner'), df_list)
df_merged.fillna('void', inplace=True)

df_merged.to_csv('Merged_HealthyAndDiseased_Top2B.csv')
print(df_merged.shape)
df_merged.head(2)
```
