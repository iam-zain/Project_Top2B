```python
import csv
import pandas as pd

%cd Z:\Project_Top2B\April_2023\GSE109627

# Define the input and output file paths
input_file = "GSE109627_MTG_processed_GA_illumina_methylation.txt"
output_file = "GSE109627.csv"

# Read the text file
with open(input_file, 'r') as f:
    lines = f.readlines()

# Create a list of lists, where each sublist represents a row in the CSV file
data = [line.strip().split() for line in lines]

# Write the data to the CSV file
with open(output_file, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
# Removing P val columns
# renaming columns

df = pd.read_csv('GSE109627.csv')
df = df[df.columns.drop(list(df.filter(regex='Pval')))]
df.columns = df.columns.str.replace('_Beta', '')
print(df.shape)
df.head(2)
```

```python
df2 = pd.read_csv('GSE109627_Age.csv')
df2.head(2)
```

```python
# Extract the names which are control
names = df2['Sample_description'].tolist()

# Subset df1 using the names extracted from df2
df1_subset = df[names]

# Adding ID_Ref column from df
df1_subset.insert(loc=0, column='ID_REF', value=df.iloc[:, 0])

df1_subset.to_csv('GSE109627_noPval.csv', index = False)

print(df1_subset.shape)
df1_subset.head(2)
```
