```python
import pandas as pd

%cd Z:\Project_Top2B\April_2023\regionwise\temporal_lobe\Diseased

df1 = pd.read_csv('GSE53162_Top2b_tl_AD.csv')
df2 = pd.read_csv('GSE76105_Top2b_tl_AD.csv')
df3 = pd.read_csv('GSE80970_Top2b_tl_AD.csv')
df4 = pd.read_csv('GSE105109_Top2b_tl_AD.csv')
df5 = pd.read_csv('GSE109627_Top2b_tl_AD.csv')
df6 = pd.read_csv('GSE125895_Top2b_tl_AD.csv')
df7 = pd.read_csv('GSE134379_Top2b_tl_AD.csv')
```

```python
df1 = df1.transpose()
df1.columns = df1.iloc[0]
df1 = df1.iloc [1:, :]

df2 = df2.transpose()
df2.columns = df2.iloc[0]
df2 = df2.iloc [1:, :]

df3 = df3.transpose()
df3.columns = df3.iloc[0]
df3 = df3.iloc [1:, :]

df4 = df4.transpose()
df4.columns = df4.iloc[0]
df4 = df4.iloc [1:, :]

df5 = df5.transpose()
df5.columns = df5.iloc[0]
df5 = df5.iloc [1:, :]

df6 = df6.transpose()
df6.columns = df6.iloc[0]
df6 = df6.iloc [1:, :]

df7 = df7.transpose()
df7.columns = df7.iloc[0]
df7 = df7.iloc [1:, :]

df7.head(2)
```

```python
from functools import reduce

df_list = [df1, df2, df3, df4, df5, df6, df7]

df_merged = reduce(lambda left, right: pd.merge(left, right, left_index=True, right_index=True, how='inner'), df_list)
df_merged.fillna('void', inplace=True)

df_merged.to_csv('Merged_Top2B_Temporal_Diseased.csv')
print(df_merged.shape)
df_merged.head(2)
```
