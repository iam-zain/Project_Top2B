```python
import csv
import pandas as pd

%cd Z:\Project_Top2B\April_2023\GSE105109

# Define the input and output file paths
input_file = "GSE105109_GEO_beta.txt"
output_file = "GSE105109.csv"

# Read the text file
with open(input_file, 'r') as f:
    lines = f.readlines()

# Create a list of lists, where each sublist represents a row in the CSV file
data = [line.strip().split() for line in lines]

# Write the data to the CSV file
with open(output_file, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
import re

# Define the input and output file paths
input_file = "GSE105109_Age.txt"
output_file = "GSE105109_Age.csv"

# Read the text file
with open(input_file, 'r') as f:
    lines = f.readlines()

# Create a list of lists, where each sublist represents a row in the CSV file
data = []
for line in lines:
    row = []
    for match in re.findall(r'"(.*?)"|(\S+)', line):
        row.append(match[0] if match[0] else match[1])
    data.append(row)

# Write the data to the CSV file
with open(output_file, 'w', newline='') as f:
    writer = csv.writer(f)
    writer.writerows(data)
```

```python
# Extract the names which are present in both df2 and df
names = list(set(df2['Sample_ID']).intersection(set(df.columns)))

# Subset df using the names extracted from df2
df_subset = df[names]

# Adding CpG column from df
df_subset.insert(loc=0, column='ID_REF', value=df.iloc[:, 0])

df_subset.to_csv('GSE105109_Control.csv', index=False)

print(df_subset.shape)
df_subset.head(2)
```
