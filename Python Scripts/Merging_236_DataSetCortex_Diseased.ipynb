```python
import pandas as pd

%cd Z:\Project_Top2B\April_2023\Horvath\From_OurModel

df1 = pd.read_csv('GSE67749_236_HIV_Age.csv')
df2 = pd.read_csv('GSE67748_236CpG_HIV_Age.csv')
df3 = pd.read_csv('GSE59457_236_HIV_Age.csv')
df4 = pd.read_csv('GSE53162_236CpG_Auti_Age.csv')
df6 = pd.read_csv('Merged_236_AD_Age_Mean.csv')
df7 = pd.read_csv('GSE74193_236CpG_Schiz_Age.csv')
```

```python
df1 = df1.transpose()
df1.columns = df1.iloc[0]
df1 = df1.iloc [1:, :]

df2 = df2.transpose()
df2.columns = df2.iloc[0]
df2 = df2.iloc [1:, :]

df3 = df3.transpose()
df3.columns = df3.iloc[0]
df3 = df3.iloc [1:, :]

df4 = df4.transpose()
df4.columns = df4.iloc[0]
df4 = df4.iloc [1:, :]

df6 = df6.transpose()
df6.columns = df6.iloc[0]
df6 = df6.iloc [1:, :]

df7 = df7.transpose()
df7.columns = df7.iloc[0]
df7 = df7.iloc [1:, :]

df7.head(2)
```

```python
from functools import reduce

df_list = [df1, df2, df3, df4, df6, df7]

df_merged = reduce(lambda left, right: pd.merge(left, right, left_index=True, right_index=True, how='inner'), df_list)
df_merged.fillna('void', inplace=True)

df_merged.to_csv('Merged_236_Diseased_Cortex.csv')
print(df_merged.shape)
df_merged.head(2)
```

```python
import numpy as np
%cd Z:\Project_Top2B\April_2023\Cortex_Clock\Our_Model

data = pd.read_csv('Merged_236_Diseased_Cortex_Age.csv')
data = data.drop(['Age'], axis = 1)

# Replace '' values with NaN
data = data.replace('', np.nan)

# Loop over each column and replace NaN values with column means
for col in data.columns:
    # Convert the column to numeric data type
    data[col] = pd.to_numeric(data[col])
    
    # Calculate the mean of the column
    col_mean = data[col].mean()
    
    # Replace NaN values with the column mean
    data[col].fillna(col_mean, inplace=True)

data.to_csv("Merged_236_Diseased_Cortex_Mean.csv", index = False)
data.head(2)
```
